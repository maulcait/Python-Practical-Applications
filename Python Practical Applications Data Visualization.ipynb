# Data Visualization Example

### Dataset comes from Kaggle 
#### https://www.kaggle.com/mirichoi0218/insurance

##### Columns
**age:** age of primary beneficiary

**sex:** insurance contractor gender, female, male

**bmi:** Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

**children:** Number of children covered by health insurance / Number of dependents

**smoker:** Smoking

**region:** the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

**charges:** Individual medical costs billed by health insurance

```python
#Import packages
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io
```

```python
#set a specific color pallete
sns.set_palette('muted')
#sns.set_palette(sns.color_palette("Set3", 10))
```

```python
# Downloading the csv file from your GitHub account

url = "https://raw.githubusercontent.com/maulcait/Python-Practice-Applications/main/insurance.csv" 
# Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
df.head()
```

```python
# Add some calculated variables to improve look of plots
df['smoker'] = np.where(df['smoker'] == 'yes', 'smoker', 'non-smoker')
```

```python
# Plot the histogram thanks to the distplot function
sns.displot(data=df, x="charges", kde=True)
```

```python

sns.displot(data=df, x="charges", kde=False, hue="smoker", multiple="stack")
```

```python
# Using seaborn to style a matpplot lib plot
# use the function regplot to make a scatterplot
sns.regplot(x=df["bmi"], y=df["charges"], fit_reg=False)
```

```python
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="bmi", y="charges", data=df, fit_reg=False, hue='smoker', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='upper left')

plt.show()
```

```python
# Basic correlogram
sns.pairplot(df)
```

```python
# Basic correlogram
sns.pairplot(df, hue="smoker")
```
