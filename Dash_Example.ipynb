```python
#Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import io
# Downloading the csv file from your GitHub account

url = "https://raw.githubusercontent.com/maulcait/Python-Practice-Applications/main/insurance.csv" 
# Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))
df.head()
```

```python
import dash
import dash_html_components as html
import plotly.graph_objects as go
import dash_core_components as dcc
import plotly.express as px
from dash.dependencies import Input, Output


app = dash.Dash()


app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Charges by BMI', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),
    
 html.Div(children='''
        Select a metric to display scatter plot by group.
    '''),
        dcc.Dropdown( id = 'dropdown',
        options = [
            {'label':'Smoker', 'value':'smoker' },
            {'label': 'Region', 'value':'region'},
            {'label': 'Gender', 'value':'sex'},
            ],
        value = 'smoker'),
        dcc.Graph(id = 'scatter')
    ])
    
@app.callback(Output(component_id='scatter', component_property= 'figure'),
              [Input(component_id='dropdown', component_property= 'value')])
def graph_update(dropdown_value):
    print(dropdown_value)
    fig = px.scatter(df, x='bmi', y="charges", color=df['{}'.format(dropdown_value)])

    return fig  



if __name__ == '__main__': 
    app.run_server()
```
